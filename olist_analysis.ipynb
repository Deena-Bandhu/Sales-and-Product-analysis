2. Yearly summary
o Create a table with one row per year showing:
1. year
2. total_orders
3. total_revenue
4. avg_order_value

```python
orders_enriched['year'] = pd.to_datetime(
    orders_enriched['order_purchase_timestamp']
).dt.year
```

```python
orders_enriched['year']
```

```python
orders_enriched.columns
```

```python
yearly_summary = (
    orders_enriched
    .groupby('year')
    .agg(
        total_orders=('order_id', 'nunique'),
        total_revenue=('payment_value', 'sum'),
        avg_order_value=('payment_value', 'mean')
    )
    .reset_index()
)
```

A4. Sales trend over time
Create a monthly aggregate table:
 Group by year_month (e.g., 2017-03), derived from order_purchase_date.
 For each year_month, compute:
1. total_orders
2. total_revenue
3. avg_order_value

Name this output table/dataframe monthly_sales.
(This will be used later in your BI dashboard.)

```python
orders_enriched.loc[:, 'year_month'] = (
    pd.to_datetime(orders_enriched['order_purchase_timestamp'])
    .dt.to_period('M')
    .astype(str)
)
```

```python
orders_enriched.year_month
```

```python
monthly_sales = (
    orders_enriched
    .groupby('year_month')
    .agg(
        total_orders=('order_id', 'nunique'),
        total_revenue=('payment_value', 'sum'),
        avg_order_value=('payment_value', 'mean')
    )
    .reset_index()
)
```

```python
monthly_sales
```

1. product_category_name

3. total_items (count of rows / items)
4. total_revenue (sum of payment_value or price – be consistent and note
what you chose)
5. avg_price (average price)

2. Sort by total_revenue (descending) and keep the top 10 categories.

```python
orders_enriched.columns
```

Grouping by Product Category

```python
category_performance = (
    orders_enriched
    .groupby('product_category_name')
    .agg(
        total_orders=('order_id', 'nunique'),
        total_items=('order_id', 'count'),
        total_revenue=('price', 'sum'),
        avg_price=('price', 'mean')
    )
    .reset_index()
)
```

2. Sort by total_revenue (descending) and keep the top 10 categories.

```python
category_performance = category_performance.sort_values(
    by='total_revenue',
    ascending=False
)
```

A6. State-level performance
Create a table state_performance with:
1. customer_state
2. unique_customers
3. total_orders
4. total_revenue
5. avg_order_value
Sort by total_revenue (descending).

```python
orders_enriched.columns
```

```python
state_performance = (
    orders_enriched
    .groupby('customer_state')
    .agg(
        unique_customers=('customer_id', 'nunique'),
        total_orders=('order_id', 'nunique'),
        total_revenue=('payment_value', 'sum'),
        avg_order_value=('payment_value', 'mean')
    )
    .reset_index()
)
```

A7. Delivery performance (basic)
Using orders table:
1. Create columns:

1. delivery_time_days = difference in days between
order_delivered_customer_date and order_purchase_timestamp.
2. estimated_delivery_days = difference in days between
order_estimated_delivery_date and order_purchase_timestamp.

2. Create a table delivery_summary_by_state:

1. customer_state
2. avg_delivery_time_days
3. avg_estimated_delivery_days
(You can join customers to get customer_state.)

```python
orders_enriched.columns
```

```python
orders_dataset
```

```python
orders_dataset.columns
```

```python
orders_dataset['order_purchase_timestamp'] = pd.to_datetime(

    orders_dataset['order_purchase_timestamp']
)
```

```python
orders_dataset['order_estimated_delivery_date'] = pd.to_datetime(

    orders_dataset['order_estimated_delivery_date']
)
```

```python
orders_dataset['order_delivered_customer_date'] = pd.to_datetime(

    orders_dataset['order_delivered_customer_date']
)
```

Creating delivery_time_days

```python
orders_dataset['delivery_time_days'] = (

    orders_dataset['order_delivered_customer_date']-
    orders_dataset['order_purchase_timestamp']
).dt.days
```

Create estimated_delivery_days

estimated_delivery_days =
order_estimated_delivery_date − order_purchase_timestamp

```python
orders_dataset['estimated_delivery_days'] = (


    orders_dataset['order_estimated_delivery_date'] -
    orders_dataset['order_purchase_timestamp']
).dt.days
```

```python
orders_dataset
```

Remove Invalid Rows

Some orders:

Are not delivered

Have missing delivery dates

We remove them to avoid wrong averages.

```python
orders_delivery = orders_dataset.dropna(

subset=['delivery_time_days', 'estimated_delivery_days']

)
```

```python
orders_delivery
```

```python
customers.columns
```

```python
orders_delivery = orders_delivery.merge(
    customers[['customer_id', 'customer_state']],
    on='customer_id',
    how='left'
)
```

```python
orders_delivery
```

```python
orders_delivery.columns
```

Grouping by State & Calculate Averages

```python
orders_delivery
```

```python
delivery_summary_by_state = (
    orders_delivery
    .groupby('customer_state')
    .agg(
        avg_delivery_time_days=('delivery_time_days', 'mean'),
        avg_estimated_delivery_days=('estimated_delivery_days', 'mean')
    )
    .reset_index()
)
```

```python
delivery_summary_by_state
```
